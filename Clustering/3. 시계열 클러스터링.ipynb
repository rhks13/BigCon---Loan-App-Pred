```python
# Native libraries
import os
import math
# Essential Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Preprocessing
from sklearn.preprocessing import MinMaxScaler
# Algorithms
from minisom import MiniSom
from tslearn.barycenters import dtw_barycenter_averaging
from tslearn.clustering import TimeSeriesKMeans
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

# **SOM - sub**

```python
df_sub.values
```

```python
som_x = som_y = 3
# I didn't see its significance but to make the map square,
# I calculated square root of map size which is 
# the square root of the number of series
# for the row and column counts of som

som = MiniSom(som_x, som_y,len(df_sub.values[0]), sigma=0.3, learning_rate = 0.1)

som.random_weights_init(df_sub.values)
som.train(df_sub.values, 10000)
```

```python
def plot_som_series_dba_center(som_x, som_y, win_map):
    fig, axs = plt.subplots(som_x,som_y,figsize=(25,25))
    fig.suptitle('Clusters')ㅇ
    for x in range(som_x):
        for y in range(som_y):
            cluster = (x,y)
            if cluster in win_map.keys():
                for series in win_map[cluster]:
                    axs[cluster].plot(series,c="gray",alpha=0.5) 
                axs[cluster].plot(dtw_barycenter_averaging(np.vstack(win_map[cluster])),c="red")
            axs[cluster].set_title(f"Cluster {cluster_number}")

    plt.show()
```

```python
# figure를 위한 라이브러리
import matplotlib as mpl
```

```python
win_map = som.win_map(df_sub.values)
# dba를 이용한 매핑
plot_som_series_dba_center(som_x, som_y, win_map)
plt.figure(dpi=600, figsize=(10, 8))
mpl.rcParams['font.family']='Arial'
mpl.rcParams['font.size']=30
```

```python
#cluster dist확인

cluster_c = []
cluster_n = []
for x in range(som_x):
    for y in range(som_y):
        cluster = (x,y)
        if cluster in win_map.keys():
            cluster_c.append(len(win_map[cluster]))
        else:
            cluster_c.append(0)
        cluster_number = x*som_y+y+1
        cluster_n.append(f"Cluster {cluster_number}")

plt.figure(figsize=(40,5))
plt.title("Cluster Distribution for SOM")
plt.bar(cluster_n,cluster_c)
plt.show()
```

```python
cluster_map = []
for idx in range(len(df_sub.values)):
    winner_node = som.winner(df_sub.values[idx])
    cluster_map.append((df_sub.index[idx],f"Cluster {winner_node[0]*som_y+winner_node[1]+1}"))

sub_cluster=pd.DataFrame(cluster_map,columns=["Series","Cluster"]).set_index("Series")
```
